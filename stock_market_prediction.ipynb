```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
url = '/content/TSLA.csv'
```

```python
df  = pd.read_csv(url)
```

```python
df
```

```python
plt.figure(figsize=(12.5, 4.5))
plt.plot(df['Adj Close'], label = 'TSLA')
plt.title('Tesla Adj. close price history')
plt.xlabel('dates mentioned')
plt.ylabel('Adj. close price')
plt.show()
```

# New Section

```python
SMA30 = pd.DataFrame()
SMA30['Adj Close Price'] = df['Adj Close'].rolling(window=30).mean()

SMA30
```

```python
# moving 100 day average 
SMA100 = pd.DataFrame()
SMA100['Adj Close Price'] = df['Adj Close'].rolling(window=100).mean()
SMA100
```

```python
plt.figure(figsize=(18.5, 7.5))
plt.plot(df['Adj Close'], label = 'TSLA')
plt.plot(SMA30['Adj Close Price'], label = 'SM30')
plt.plot(SMA100['Adj Close Price'], label = 'SM100')
plt.title('Tesla Adj. close price history')
plt.xlabel('dates mentioned')
plt.ylabel('Adj. close price')
plt.legend(loc='upper left')
plt.show()
```
